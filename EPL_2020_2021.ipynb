```python
import pandas as pd #Data manipulation
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset
epl_df = pd.read_csv('EPL_20_21.csv')
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
epl_df.isna().sum()
```

```python
# Create two new columns
epl_df['MinsperMatch'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['GoalsperMatch'] = (epl_df['Goals'] / epl_df['Matches']).astype(float).round(2)
epl_df.head()
```

```python
# Total Goals
Total_goals = epl_df['Goals'].sum()
Total_goals
```

```python
# Penalty goals
Penalty_goals = epl_df['Penalty_Goals'].sum()
Penalty_goals
```

```python
# Penalty Attempts
penalty_attempts = epl_df['Penalty_Attempted'].sum()
penalty_attempts
```

```python
# Pie chart for penalties scored and missed goals
plt.figure(figsize= (14,7))
pl_notscored = epl_df['Penalty_Attempted'].sum() - Penalty_goals
data = [pl_notscored, Penalty_goals]
labels = ['Penalties Missed', 'Penalties scored']
color = sns.color_palette('tab10')
plt.pie(data, labels= labels, colors= color, autopct= '%.0f%%')
plt.show()
```

```python
# Unique Postions
epl_df['Position'].unique()
```

```python
# Total FW Players
epl_df[epl_df['Position'] == 'FW']
```

```python
# Players from different Nations
np.size((epl_df['Nationality'].unique()))
```

```python
# Most players from which Countries
nationality= epl_df.groupby('Nationality').size().sort_values(ascending= False)
nationality.head(10).plot(kind = 'bar', figsize= (12,6), color = sns.color_palette('magma'))
```

```python
# Clubs with max. players in thier squad
epl_df['Club'].value_counts().nlargest(6).plot(kind= 'bar', color = sns.color_palette('viridis'))
```

```python
# Clubs with min. players in thier squad
epl_df['Club'].value_counts().nsmallest(6).plot(kind= 'bar', color = sns.color_palette('viridis'))
```

```python
# Players based on Age groups
Under20 = epl_df[epl_df['Age'] <= 20]
age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
Above30 = epl_df[epl_df['Age'] > 30]
```

```python
x = np.array([Under20['Name'].count(), 
              age20_25['Name'].count(), 
              age25_30['Name'].count(),
              Above30['Name'].count()])
mylabels = ['<=20', '>20 & <=25', '<25 & <=30', '>30']
plt.title('Total Players with Age', fontsize = 20)
plt.pie(x, labels = mylabels, autopct= '%.1f%%', center=(0, 0), pctdistance=0.8)
plt.show()
```

```python
players_under_20 = epl_df[epl_df['Age'] < 20]
players_under_20['Club'].value_counts().plot( kind = 'bar', figsize= (12,6), color = sns.color_palette('cubehelix'))
```

```python
# Under 20 players in MaNU
players_under_20[players_under_20['Club'] == 'Manchester United']
```

```python
# Under 20 players in Chelsea
players_under_20[players_under_20['Club'] == 'Chelsea']
```

```python
# Avg age of players
plt.figure(figsize= (12,6))
sns.boxenplot(x = 'Club', y = 'Age', data = epl_df)
plt.xticks(rotation = 90)
plt.show()
```

```python
num_player = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum()) / num_player
data.sort_values(ascending= False)
```

```python
# Total assists from each CLUB
Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index= False)['Assists'].sum())
sns.set_theme(style= 'whitegrid', color_codes= True)
ax = sns.barplot(x = 'Club', y= 'Assists', data = Assists_by_clubs.sort_values(by='Assists'), palette= 'Set2')
ax.set_xlabel('Club', fontsize = 20)
ax.set_ylabel('Assists', fontsize = 20)
plt.xticks(rotation = 75)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of the Clubs vs Total assists', fontsize= 24)
```

```python
# Top 10 Assist
top_10_assists = epl_df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns = 'Assists')
top_10_assists
```

```python
Goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index= False)['Goals'].sum())
sns.set_theme(style= 'whitegrid', color_codes= True)
ax = sns.barplot(x = 'Club', y= 'Goals', data = Goals_by_clubs.sort_values(by='Goals'), palette= 'rocket')
ax.set_xlabel('Club', fontsize = 30)
ax.set_ylabel('Goals', fontsize = 20)
plt.xticks(rotation = 75)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of the Clubs vs Total GOALS', fontsize= 20)
```

```python
# Most Goals by players
top_10_goals = epl_df[['Name', 'Club', 'Goals', 'Matches']].nlargest(n=10, columns= 'Goals')
top_10_goals
```

```python
# Goals per match
top_10_goalspermatch = epl_df[['Name', 'GoalsperMatch', 'Matches', 'Goals']].nlargest(n=10, columns = 'GoalsperMatch')
top_10_goalspermatch
```

```python
# Pie chart - Goals with assist and without assists
plt.figure(figsize= (14,7))
assists = epl_df['Assists'].sum()
data = [Total_goals - assists, assists]
labels = ['Goals without Assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data, labels= labels, colors = color, autopct= '%.0f%%')
plt.show()
```

```python
# Top 10 player with most Yellow cards

epl_yellow = epl_df.sort_values(by='Yellow_Cards', ascending= False)[:10]
plt.figure(figsize= (20,6))
plt.title('Players with Most Yellow cards')
c = sns.barplot(x = epl_yellow['Name'], y = epl_yellow['Yellow_Cards'], label = 'Players', color = 'yellow')
plt.ylabel('Number of Yellow cards')
c.set_xticklabels(c.get_xticklabels(), rotation= 45)
c
```
